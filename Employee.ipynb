## Feature Engineering

```python
# Using the preprocessed data
df = pd.read_csv('Preprocessed.csv')
df
```

```python
# Drop Unnamed
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# Engagement Factors
engagement_factors = [
    'JobInvolvement',
    'JobSatisfaction',
    'EnvironmentSatisfaction',
    'WorkLifeBalance',
    'RelationshipSatisfaction'
]

df['EngagementScore'] = df[engagement_factors].mean(axis=1).astype(int)
```

```python
df['EngagementScore']
```

```python
df['YearsAtCompany']
```

```python
# Roles and Performance
df['PromotionFrequency'] = (df['YearsAtCompany'] / (df['YearsSinceLastPromotion'] + 1)).round(2).astype(int)
df['RoleChangeFrequency'] = (df['YearsAtCompany'] / (df['YearsInCurrentRole'] + 1)).round(2).astype(int)
```

```python
# IncomePerYearOfService
df['IncomePerYearOfService'] = ((df['MonthlyIncome']*12)/df['YearsAtCompany']+1).round(2)
df['IncomePerYearOfService']
```

```python
np.isinf(df).sum()
```

```python
# IncomePerYearOfService contains the nan,inf values
df.replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
np.isinf(df).sum()
```

```python
df['IncomePerYearOfService'] = df['IncomePerYearOfService'].astype(int)
```

```python
# Monthly income by Age
df['IncomeToAgeRatio'] =  (df['MonthlyIncome'] / df['Age']).round(2).astype(int)
df['IncomeToAgeRatio']
```

```python
# MonthlyIncome by YearsSinceLastPromotion
df['PromotionToIncomeGrowth'] = (df['MonthlyIncome'] / (df['YearsSinceLastPromotion'] + 1)).round(2).astype(int)
df['PromotionToIncomeGrowth']
```

```python
# Work-Life Balance
df['OverallSatisfaction'] = ((df['EnvironmentSatisfaction'] + df['JobSatisfaction'] + df['JobInvolvement']) / 3).round(2).astype(int)
df['OverallSatisfaction']
```

```python
df.drop(['EnvironmentSatisfaction','JobSatisfaction','JobInvolvement'],inplace=True,axis=1)
```

```python
# Create the AgeGroup category
bins = [18, 30, 40, 50, 66]
labels = ['18-30', '31-40', '41-50', '51-65']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# One-hot encode the new column and update the DataFrame
df = pd.get_dummies(df, columns=['AgeGroup'], drop_first=True)
```

```python
df = df.drop('Age', axis=1)
```

```python
# YearsAtCompany

df['TenureCategory'] = pd.cut(df['YearsAtCompany'],
                              bins=[-1, 2, 5, 10, df['YearsAtCompany'].max()],
                              labels=['0-2 years', '2-5 years', '5-10 years', '10+ years'])

# Now, perform the one-hot encoding and update 'df' in place
df = pd.get_dummies(df, columns=['TenureCategory'], drop_first=True)
```

```python
df = df.drop('YearsAtCompany', axis=1)
```

```python
df.head()
```

```python
for col in df.columns:
    if df[col].dtype == 'bool':
        df[col] = df[col].astype(int)
```
